```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import pickle

# Separate features and target
X_heart = df_heart.drop(columns=['target'])
y_heart = df_heart['target']

print("=== Features ===")
print(X_heart.columns.tolist())

# Split into train and test
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(
    X_heart, y_heart,
    test_size=0.2,
    random_state=42
)

# Scale the data
scaler_heart = StandardScaler()
X_train_h_scaled = scaler_heart.fit_transform(X_train_h)
X_test_h_scaled = scaler_heart.transform(X_test_h)

# Train Random Forest model
heart_model = RandomForestClassifier(random_state=42, n_estimators=100)
heart_model.fit(X_train_h_scaled, y_train_h)

# Test accuracy
y_pred_h = heart_model.predict(X_test_h_scaled)
accuracy = accuracy_score(y_test_h, y_pred_h) * 100
print(f"\n=== Heart Disease Model Accuracy ===")
print(f"Random Forest: {accuracy:.2f}%")

print("\n=== Detailed Report ===")
print(classification_report(y_test_h, y_pred_h,
      target_names=['No Heart Disease', 'Heart Disease']))
```

```python
# Save heart model and scaler
with open('heart_model.pkl', 'wb') as f:
    pickle.dump(heart_model, f)

with open('heart_scaler.pkl', 'wb') as f:
    pickle.dump(scaler_heart, f)

print("Heart model and scaler saved successfully!")
```
